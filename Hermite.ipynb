```python
from scipy.special import eval_hermite

def get_N(n):
    return 1./np.sqrt(np.sqrt(np.pi)*2**n*np.math.factorial(n))

x = np.linspace(-5,5,80)

n = 20

h = eval_hermite(n,x)

H = get_N(n) * h * np.exp(-0.5*x**2)


plt.plot(x,H)
plt.show()
```
